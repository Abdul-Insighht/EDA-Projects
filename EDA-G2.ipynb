```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the dataset
df = pd.read_csv('Suicide_bombing_attacks.csv')
```

```python
# Data Cleaning
# Handle missing values - you may choose to fill or drop based on the context
df.dropna(subset=['Location', 'Date'], inplace=True)  # Drop rows with essential missing values

# Convert 'Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Extract year and month for furtheranalysis
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()
```

```python
# 1. Distribution of Attacks Over the Years
plt.figure(figsize=(12, 6))
sns.countplot(x='Year', data=df, palette='viridis')
plt.title('Number of Suicide Bombing Attacks Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Monthly Trend of Attacks
plt.figure(figsize=(12, 6))
sns.countplot(x='Month', data=df, palette='Set2', order=[
    'January', 'February', 'March', 'April', 'May', 'June', 'July',
    'August', 'September', 'October', 'November', 'December'])
plt.title('Monthly Distribution of Suicide Bombing Attacks')
plt.xlabel('Month')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Locations of Attacks
plt.figure(figsize=(12, 8))
sns.countplot(y='Location', data=df, order=df['Location'].value_counts().index, palette='coolwarm')
plt.title('Suicide Bombing Attacks by Location')
plt.xlabel('Number of Attacks')
plt.ylabel('Location')
plt.show()
```

```python
# 4. Severity of Attacks (e.g., casualties)
if 'Casualties' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.histplot(df['Casualties'], bins=30, kde=True, color='red')
    plt.title('Distribution of Casualties in Suicide Bombing Attacks')
    plt.xlabel('Number of Casualties')
    plt.ylabel('Frequency')
    plt.show()
```

```python
# 5. Relationship between Year and Casualties
if 'Casualties' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='Year', y='Casualties', data=df)
    plt.title('Casualties per Year')
    plt.xlabel('Year')
    plt.ylabel('Number of Casualties')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 6. Heatmap of Attacks Over Years and Locations
attack_counts = df.pivot_table(index='Year', columns='Location', aggfunc='size', fill_value=0)
plt.figure(figsize=(15, 10))
sns.heatmap(attack_counts, annot=True, fmt='d', cmap='Blues')
plt.title('Heatmap of Suicide Bombing Attacks Over Years and Locations')
plt.xlabel('Location')
plt.ylabel('Year')
plt.show()
```

```python
# 7. Trend Analysis: Yearly Casualties (if applicable)
if 'Casualties' in df.columns:
    yearly_casualties = df.groupby('Year')['Casualties'].sum().reset_index()
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=yearly_casualties, x='Year', y='Casualties', marker='o')
    plt.title('Yearly Casualties from Suicide Bombing Attacks')
    plt.xlabel('Year')
    plt.ylabel('Total Casualties')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 8. Save cleaned data for further analysis
df.to_csv('Cleaned_Suicide_Bombing_Attacks.csv', index=False)

print("\nEDA completed successfully.")
```
