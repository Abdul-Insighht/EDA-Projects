```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure only numerical columns are selected
numerical_columns = df.select_dtypes(include=['float64', 'int64'])

# Customize outliers' appearance (flierprops)
flierprops = dict(marker='o', markerfacecolor='red', markersize=8, linestyle='none')

# Increase the size of the plot for better visibility
plt.figure(figsize=(16, 10))

# Create a boxplot for all numerical columns, highlighting the outliers in red
sns.boxplot(data=numerical_columns, orient='h', flierprops=flierprops)

# Add title and rotate y-axis labels (if needed)
plt.title('Box Plot for Numerical Columns with Highlighted Outliers')
plt.show()
```

```python
# Data Cleaning
cleaned_data = df.dropna()  # Remove missing values
cleaned_data = cleaned_data.drop_duplicates()  # Remove duplicates
```

```python
import calendar  # import Calendar module that use to handle month names and numbers 
# Convert month names to numbers for seasonality analysis
cleaned_data['arrival_date_month_num'] = cleaned_data['arrival_date_month'].apply(lambda x: list(calendar.month_name).index(x))
```

```python
# 1. Seasonality Analysis (Bookings per month and year)
seasonality_data = cleaned_data.groupby(['arrival_date_year', 'arrival_date_month_num']).size().reset_index(name='num_bookings')
```

```python
#for ignoring all type of warnings
import warnings

# Ignore all warnings
warnings.filterwarnings("ignore")
```

```python
# Plot seasonality trends
plt.figure(figsize=(12, 6))
sns.lineplot(x='arrival_date_month_num', y='num_bookings', hue='arrival_date_year', data=seasonality_data, marker='o')
plt.title('Hotel Bookings Seasonality (Monthly and Yearly)', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Bookings', fontsize=12)
plt.xticks(range(1, 13), calendar.month_abbr[1:])
plt.legend(title="Year")
plt.show()
```

```python
# 2. Lead Time vs Cancellations
plt.figure(figsize=(12, 6))
sns.histplot(data=cleaned_data, x='lead_time', hue='is_canceled', multiple='stack', bins=50, kde=True)
plt.title('Lead Time vs Cancellations', fontsize=16)
plt.xlabel('Lead Time (Days)', fontsize=12)
plt.ylabel('Number of Bookings', fontsize=12)
plt.show()
```

```python
# 3. Customer Preferences (Customer type vs Special Requests)
plt.figure(figsize=(12, 6))  # Plot ka size define kar rahe hain
cleaned_data['total_of_special_requests'] = cleaned_data['total_of_special_requests'].astype(str)  # Special requests column convert into string

# Ab seaborn countplot ko categorical data milega hue ke liye
sns.countplot(x='customer_type', hue='total_of_special_requests', data=cleaned_data)  

# Plot ka title, x-axis, aur y-axis labels add kar rahe hain
plt.title('Customer Type vs Special Requests', fontsize=16)  
plt.xlabel('Customer Type', fontsize=12)  
plt.ylabel('Number of Special Requests', fontsize=12)  

plt.legend(title='Special Requests')  
plt.show()  
```

```python
# 4. Pricing Analysis (ADR based on different conditions)
plt.figure(figsize=(12, 6))
sns.boxplot(x='customer_type', y='adr', data=cleaned_data)
plt.title('Price Distribution (ADR) by Customer Type', fontsize=16)
plt.xlabel('Customer Type', fontsize=12)
plt.ylabel('Average Daily Rate (ADR)', fontsize=12)
plt.show()
```
